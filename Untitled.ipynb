```python
%matplotlib inline
train_data['SALE PRICE'] = train_data['SALE PRICE'].astype(long)
train_data.head()
observe_data = train_data['SALE PRICE'].apply(lambda x : x % 1000)
observe_data__means = observe_data.mean()
observe_data__std= observe_data.std()
std_data = (observe_data - observe_data__means) / observe_data__std
prices = pd.DataFrame({"Price":observe_data, "Log(Price + 1)":std_data})
prices.hist(figsize = (10,10),xlabelsize=20)
```

```python
#标准化numerical数据
#因为regression 分类器比较傲娇，最好是把源数据放在一个标准分布里，不要让数据差距太大
numerical_cols = all_data.columns[all_data.dtypes != 'object']
numerical_cols
```

```python
#计算标准分布，让数据更加平滑，有利于计算
numerical_cols_means = all_dummy_data.loc[:,numerical_cols].mean()
numerical_cols_std = all_dummy_data.loc[:,numerical_cols].std()
all_dummy_data.loc[:,numerical_cols] = (all_dummy_data.loc[:,numerical_cols] - numerical_cols_means) / numerical_cols_std
```

模型的建立

```python
dummy_train_data = all_dummy_data.loc[train_data.index]
dummy_test_data = all_dummy_data.loc[test_data.index]
```

```python
dummy_train_data.shape, dummy_test_data.shape
```

```python
x_train = dummy_train_data[0:10001].values
x_test = dummy_test_data[0:10001].values
y_train = y_train[0:10001]
```

```python
#我们先选择使用Ridge Regression
#因为队友多音字数据，这种模型可以方便的把所有的var放进去
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
np.isnan(x_train).sum().sort_values()
```

```python
#用Sklearn自带的cross validation方法来测试模型
alphas = np.logspace(-3,2,50)
test_scores = []
for alpha in alphas:
    clf = Ridge(alpha)
    test_score = np.sqrt(-cross_val_score(clf, x_train, y_train, cv = 10, scoring = 'neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
#存下所有的CV值,看看哪个alpha值更好
plt.plot(alphas, test_scores)
plt.title("Alpha vs CV Error")
```
